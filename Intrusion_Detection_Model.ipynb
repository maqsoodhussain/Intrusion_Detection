```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
```

importing dataset  and print first 5 rows 

```python
df = pd.read_csv('data/Train_data.csv')
df.head()
```

Check unique values in protocol_type feature

```python
df['protocol_type'].unique()
```

```python
proto_dict = {
    'tcp':0,
    'udp':1,
    'icmp':2
}

df['protocol_type']=df['protocol_type'].map(proto_dict)

service_to_number  = {service: i for i , service in enumerate(df['service'].unique())}
df['service'] = df['service'].map(service_to_number)


flag_to_number  = {flag: i for i , flag in enumerate(df['flag'].unique())}
df['flag'] = df['flag'].map(flag_to_number)
```

```python
df.head()
```

```python
df['flag'].unique()
```

```python
df['class'].unique()
```

```python

class_dict  = {
    'normal':0,
    'anomaly':1
}

df['class'] = df['class'].map(class_dict)
df.head()
```

```python
df.head()
```

```python
df.dropna(axis=0)
```

Check isany null value present or not 

```python
# empty_row = df.isnull().all(axis=0)
# empty_col= df.isnull().all(axis=1)
# print(empty_row)
# print(empty_col)
```

```python
y = df['class']
print(y)
```

```python
X = df.drop('class', axis=1)
print(X)
```

```python
Train_X, Val_x, Train_y, Val_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
df_model = RandomForestClassifier()
df_model.fit(Train_X,Train_y)

prediciton = df_model.predict(Val_x)
```

```python
print(prediciton)
```

```python
print(classification_report(Val_y,prediciton))
```

```python
print(accuracy_score(Val_y,prediciton))
```
